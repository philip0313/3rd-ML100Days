```python
df=pd.DataFrame(app_train)
print(df)
```

```python
df['AMT_INCOME_TOTAL']
print(df['AMT_INCOME_TOTAL'].mean())#平均數
print(df['AMT_INCOME_TOTAL'].std())#標準差
print(df['AMT_INCOME_TOTAL'].max())#最大
print(df['AMT_INCOME_TOTAL'].min())#最小
```

```python
df['AMT_INCOME_TOTAL'].mean()
df['AMT_INCOME_TOTAL'].std()
```

```python
plt.hist(df['AMT_INCOME_TOTAL'],bins=[50000,100000,150000,200000,250000])#畫直方圖，bins可自己設定區間
plt.show()
```
